```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy.special
```

# Idea:

I am wondering how to model the allelic frequencies with a binomial process in which we pick allelic variants from an imaginary bowl of balls ..

1. Let's imagine we know the real allelic frequency: How likely is it to find it with varying sample sizes (number of times we draw from the bowl) and AL

```python
# 1. The allelic frequency is p, the number of times we draw if N, what is probability distribution for the different events?

p=[0.05,0.1,0.2,0.3,0.5,0.9] # allelic frequency of allele A at a locus 

N=[10,20,50,70,100,200]

def binomial(p,N):
    pDist = []
    for i in range(N+1):
        pDist.append(scipy.special.comb(N, i, exact=True)*p**i*(1-p)**(N-i))
    return np.array(pDist)
#scipy.special.comb(N, i, exact=True)* # p**i*(1-p)**(N-i)
```

```python
# plot a combination of values
fig, ax1 = plt.subplots()

for prob in p:
    plt.plot(list(range(N[0]+1)),binomial(prob,N[0]),"--*",label="N="+str(N[0]) + ", p= "+ str(prob))

ax1.legend()
ax1.set_xlabel('number of successes in N trials')
ax1.set_ylabel('probability')
```

```python
# plot a combination of values
fig, ax2 = plt.subplots(figsize=(12, 5))

for nn in N:
    plt.plot(list(range(nn+1)),binomial(p[0],nn),"--*",label="N="+str(nn) + ", p= "+ str(p[0]))

ax2.legend()
ax2.set_xlim([-2,20])
```

```python
# plot a combination of values
fig, ax2 = plt.subplots(figsize=(12, 5))

for nn in N:
    plt.plot(list(range(nn+1)),binomial(p[1],nn),"--*",label="N="+str(nn) + ", p= "+ str(p[1]))

ax2.legend()
ax2.set_xlim([-2,40])
```
